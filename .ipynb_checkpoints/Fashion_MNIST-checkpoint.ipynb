## Fashion MNIST Classification Project

```python
## Load libraries and dataset
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.callbacks import TensorBoard
import time
```

```python
(train_images, train_labels), (test_images, test_labels) = keras.datasets.fashion_mnist.load_data()
```

```python
# Let's view the first item!
plt.figure()
plt.imshow(train_images[0], cmap=plt.cm.binary)
plt.show()
```

```python
# Normalize and reshape
train_images = train_images / 255.0
test_images = test_images / 255.0

train_images = train_images.reshape((train_images.shape[0], 28, 28, 1))
test_images = test_images.reshape((test_images.shape[0], 28, 28, 1))
```

```python
# Let's add layers incrementally
model = keras.models.Sequential()
model.add(keras.Input(shape=(28, 28, 1)))
model.summary()
```

```python
model.add(layers.Conv2D(32, (3,3), activation="relu"))
model.summary()
```

```python
model.add(layers.MaxPooling2D(2))
model.summary()
```

```python
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.summary()
```

```python
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.summary()
```

```python
model.add(layers.Flatten())
model.summary()
```

```python
model.add(layers.Dense(32, activation='relu'))
model.summary()
```

```python
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
x_val = train_images[-10000:]
y_val = train_labels[-10000:]
x_train = train_images[:-10000]
y_train = train_labels[:-10000]
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model_1 = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_data=(x_val, y_val))
```

```python
# Plot training & validation accuracy values
plt.plot(model_1.history['accuracy'])
plt.plot(model_1.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(model_1.history['loss'])
plt.plot(model_1.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

model_2 = model.fit(x_train, y_train, epochs=50, batch_size=64, validation_data=(x_val, y_val), callbacks=[early_stopping])
```

```python
# Plot training & validation accuracy values
plt.plot(model_2.history['accuracy'])
plt.plot(model_2.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(model_2.history['loss'])
plt.plot(model_2.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
# Predict the classes for test data
predictions = model.predict(test_images)
predicted_classes = np.argmax(predictions, axis=1)

# Generate confusion matrix
cm = confusion_matrix(test_labels, predicted_classes)
print(cm)

# Generate classification report
report = classification_report(test_labels, predicted_classes)
print(report)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

```python
predictions = model.predict(test_images)

# To visualize the first prediction and the corresponding image:
print("Predicted label:", predictions[0].argmax())
print("Actual label:", test_labels[0])

plt.figure()
plt.imshow(test_images[0].reshape(28, 28), cmap=plt.cm.binary)
plt.show()
```

### Transfer Learning using Pre-trained Models

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
import numpy as np
from tensorflow.keras.applications.resnet50 import ResNet50
```

```python
# reshape the data
X_train = x_train.reshape((x_train.shape[0], 28, 28))
X_test = test_images.reshape((test_images.shape[0], 28, 28))
# change the type to float
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# convert data to 3 channels
X_train = np.stack((X_train,)*3, axis=-1)
X_test = np.stack((X_test,)*3, axis=-1)

y_train = to_categorical(y_train)
y_test = to_categorical(test_labels)

x_val = X_train[-10000:]
y_val = y_train[-10000:]
x_train = X_train[:-10000]
y_train = y_train[:-10000]
```

```python
# data augmentation with generator
#train_generator = ImageDataGenerator(
 #   rescale = 1./255,  # normalization of images
  #  rotation_range = 40, # augmention of images to avoid overfitting
   # shear_range = 0.2,
    #zoom_range = 0.2,
    #fill_mode = 'nearest'
#)

#val_generator = ImageDataGenerator(rescale = 1./255)

#train_iterator = train_generator.flow(X_train, y_train, batch_size=64, shuffle=True)

#val_iterator = val_generator.flow(X_test, y_test, batch_size=64, shuffle=False)
```

```python
model = keras.models.Sequential()
# add the pretrained model
model.add(ResNet50(include_top=False, pooling='avg', weights='imagenet'))
# add fully connected layer with output
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# set resnet layers not trainable
model.layers[0].trainable=False
model.summary()
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model.fit(x_train, y_train, epochs=50, batch_size=64, validation_data=(x_val, y_val), callbacks=[early_stopping])
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print('\nTest accuracy:', test_acc)
```
